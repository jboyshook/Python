Regular strings are those strings which can be generated by a set of rules

In web scraping regex is used in the identifying email adresses from a web page

The regex for this is 

[A-Za-z0-9\._+]+@[A-Za-z]+\.(com|org|edu|net)

The first box brackets check for the letter from A to z and from 0 to 9 then the @ symbol then the domain name

REGULAR EXPRESSIONS AND BEAUTIFUL SOUP

Lets say we want to grab the srcs of all the gift images we coud use findAll('img') however it would get all the uneccesary imgs as well

```python

from urllib.request import urlopen
from bs4 import BeautifulSoup
import regex as re

html = urlopen(
    'http://www.pythonscraping.com/pages/page3.html'
)

bsObj = BeautifulSoup(html, 'html.parser')

images = (bsObj.find_all(
    'img', {'src':re.compile('\.\.\/img\/gifts/img.*\.jpg')}
))

for image in images:
    print(image['src'])
    
```
